```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt 
```

```python
IMAGE_SIZE = 256
BATCH_SIZE = 32
CHANNEL = 3
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
           "PlantVillage", 
    shuffle = True, 
    image_size = (IMAGE_SIZE, IMAGE_SIZE),
    batch_size = BATCH_SIZE
)
```

```python
class_names = dataset.class_names
```

```python
class_names
```

```python
len(dataset)
```

```python
for image_batch, label_batch in dataset.take(1):
    print (image_batch.shape)
    print (label_batch.numpy())
    
    #the numbers in the array indicate that we have three diffrent types of images [0,1,2]
```

```python
#To print the first element [0]

for image_batch, label_batch in dataset.take(1):
    print (image_batch[0])
```

```python
#It is tensor, so we have to change into two dimensional array, we use the function numpy()

for image_batch, label_batch in dataset.take(1):
    print (image_batch[0].numpy())
```

```python
#shape of the first image 

for image_batch, label_batch in dataset.take(1):
    print (image_batch[0].shape)
```

```python
for image_batch, label_batch in dataset.take(1):
    
    plt.implot(image_batch[0].numpy().astype('uint8'))
    plt.title (label_batch[0])
    plt.axis ("off")
```
